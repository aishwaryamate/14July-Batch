# Cumulative Distribution Function(CDF)

- The CDF of a random variable 'X' may be defined as the probability that the random variable 'X' takes a value "less than or equal to x".



```python
from scipy import stats
```

```python
stats.norm.cdf(72,68,3)
```

```python
1 - stats.norm.cdf(72,68,3)
```

```python
stats.norm.cdf(70,60,10)
```

```python
1 - stats.norm.cdf(70,60,10)
```

```python
stats.norm.cdf(680,711,29)
```

```python
1- stats.norm.cdf(680,711,29)
```

```python
stats.norm.cdf(740,711,29)
```

```python
stats.norm.cdf(697,711,29)
```

# Probability Questions:

1. According to the survey, the average weight of child is 70 pounds in US with Std 10 pounds.
    - What is the probability that the weight is less than 50?
    - What is the probabilty that the weight is in range of 50 to 90?
    - What is the probabilty that the weight is greater than 100 pounds?
    - What is the probability that the weight is between 55 and 70?
    - What is the probability that the weight is equal to 75 pounds?
  

```python
stats.norm.cdf(50,70,10)
```

```python
stats.norm.cdf(90,70,10) - stats.norm.cdf(50,70,10)
```

```python
1 - stats.norm.cdf(100,70,10)
```
