# BOXPLOT

-  A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”). It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.

**COMPARISON BETWEEN BOXPLOT AND NORMAL DISTRIBUTION PLOT**

# Subplot & Subplots

```python
plt.subplot(2,2,1)
sns.boxplot(df['highway-mpg'])
plt.title('Outlier detection',size = 25)

plt.subplot(2,2,2)
sns.scatterplot(x = df['engine-size'],y = df['price'])
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.title('Scatter plot',size = 25)

plt.subplot(2,2,3)
sns.histplot(df['price'])

plt.subplot(2,2,4)
df['body-style'].value_counts().plot(kind = 'pie', autopct = '%0.0f%%')

plt.tight_layout()
```

```python
fig, index = plt.subplots(2,2)

sns.histplot(df['width'] ,ax = index[0,0])
sns.countplot(x=df['body-style'], ax=index[0,1])
sns.distplot(df['engine-size'], ax=index[1,0])
sns.scatterplot(df['width'], df['price'], ax = index[1,1])
plt.tight_layout()
```
