# Why to visualize data?
- Data visualization is the practice of translating information into a visual context, such as a map or graph, to make data easier for the human brain to understand and pull insights from. The main goal of data visualization is to make it easier to identify patterns, trends and outliers in large data sets.
    - Better analysis
    - Quick action
    - Identifying patterns
    - Finding errors
    - Understanding the story
    - Exploring business insights
    - Grasping the Latest Trends
    
# What is Matplotlib?

- Matplotlib is a python visualisation library, majorly used for plotting 2D data.
- Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack.
- Matplotlib consists of several plots like line, bar, scatter, histogram etc.

```python
##Import library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# **Matplotlib Line plot**

```python
x = np.linspace(1,20,25)
y = np.random.random_sample(25)
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.figure(figsize=(20,20))
plt.plot(x,y)
```

```python
#Set graph parameters
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 500
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color = 'red', linestyle = '--')
```

```python
#Multiline graph
```

```python
x = np.arange(1,21)

log = np.log(x)
sqrt = np.sqrt(x)

plt.plot(x,log,label ='Log')
plt.plot(x,sqrt, label = 'Sqrt')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('LineGraph')
plt.legend()
```

# Matplotlib Histogram 
- A histogram is a graph showing frequency distributions.

- It is a graph showing the number of observations within each given interval.

```python
#Read external file
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/cars.csv')
df
```

```python
plt.hist(df['width'])
```

```python
df['width'].skew()
```

```python
df.hist()
plt.tight_layout()
```

```python
df['engine-size'].skew()
```

```python
#median and mean
```

```python
mean = df['width'].mean()
median = df['width'].median()
```

```python
plt.hist(df['width'], edgecolor= 'black')
plt.axvline(mean, color = 'red')
plt.axvline(median, color = 'green')
plt.title('Frequency graph for width column')
```
